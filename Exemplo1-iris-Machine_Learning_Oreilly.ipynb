# Preparação inicial

### Importando dados

```python
from sklearn.datasets import load_iris 
iris_dataset = load_iris()
print(iris_dataset.keys(), '\n')
print(iris_dataset['DESCR'][:193] + "\n____________________________")
```

### Entendendo o Dataset:

Nós procuramos inferir o valor da feature "target_names" no nosso modelo:

```python
print(f"Target names: {iris_dataset['target_names']}")
```

E agora, dando uma olhada geral em todas as chaves do dataset para entendermos com o que estamos trabalhando: 

```python
for key in iris_dataset.keys():
    if (key != 'data' and key != 'DESCR'):
        print(f'{key} : {iris_dataset[key]} \n')
```

No elemento "data" temos os dados descritos pelo 'feature_names' indicando as dimensões das pétalas e sépalas

```python
print(iris_dataset['data'])
print(len(iris_dataset['data']), type(iris_dataset['data']))
```

Já nossa coluna 'target' nos indica o resultado de flor de cada um dos elementos da np.array 'data' já rotulados conforme a posição de 'target_names'

```python
print(iris_dataset['target'])
print(len(iris_dataset['target']), type(iris_dataset['target']))
```

ou seja: 0 - setosa, 1 - versicolor, and 2 - virginica.

# Mensurando o sucesso do projeto:

Como modelo supervisionado, vamos separar os dados num traning set e um test set. Das 150 flores, 75% delas será direcionada para treinar o algoritmo e 25% para testar e comparar o resultado da máquina com o resultado tabelado

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(
        iris_dataset['data'], iris_dataset['target'], random_state=0)

print("X_train shape: {}".format(X_train.shape)) 
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape)) 
print("y_test shape: {}".format(y_test.shape))
```

# Analisando o Dataset antes da criação do modelo:

Antes da criação do modelo, vamos verificar o dataset procurando por correlações, anormalidades e insights para nosso modelo:

```python
import pandas as pd
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
display(iris_dataframe)
```

```python
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(10, 10), marker='o',
                            hist_kwds={'bins': 20}, s=30, alpha=.8, cmap='tab20c')
```

Nesse caso, podemos ver com os dados que os tipos de flores por pétala estão bem separados. Então, a princípio, ML conseguiria prever uma nova flor sem problemas usando o algorítmo KNN

# O que é KNN e sua implementação:

KNN (K-Nearest Neighbors) é um algorítmo de ML supervisionado baseado na relação entre dados numéricos afim de prever uma variável categórica.

***Exemplo usado:*** podemos ver nos gráficos de dispersão que, quando comparamos as dimensões de sépalas e pétalas das flores, temos uma clusterização da variável categórica espécie de flor (que queremos prever)... Portanto, ao inserir um novo dado, o algorítmo KNN usa o conceito de distância euclidiana entre os pontos para definir em qual espécie o novo input se encaixa através dos K-ésimos pontos mais próximos.

```python
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors=1)
```

O módulo ***KNeighborsClassifier***  do Scikit-Learn importa esse algoritmo já implementado, encapsulando o todos os elementos do KNN na instanciação da biblioteca. Estamos instanciando o primeiro vizinho mais próximo para definir a espécie do novo input.

```python
knn.fit(X_train, y_train)
```

Com o código acima, nosso modelo definido na variável 'knn' está treinado!

# Hora das previsões!

Agora, criaremos um novo input de exemplo e o KNN criado retornará a previsão para aquele dado

```python
import numpy as np

X_new = np.array([[5, 2.9, 1, 0.2]]) #dado a ser previsto
prediction = knn.predict(X_new) 

print(f"Prediction: {prediction}") 
print(f"Predicted target name: {iris_dataset['target_names'][prediction]}")
```

Ihaaa, temos nosso modelo pronto, treinado e funcionando!

# Agora vamos validar o modelo e entender sua precisão

Como estamos tratando de um modelo supervisionado, essa validação é muito mais fácil, precisamos apenas fazer predições para os 25% dos dados que separamos inicialmente e comparar com o tabelado !

```python
y_pred = knn.predict(X_test)
print(f"Test set predictions:\n {y_pred}")
```

Usando o método 'score' do objeto 'knn' criado através do módulo KNeighborsClassifier

```python
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```

# Conclusão

Podemos ver que a implementação do algoritmo KNN não é um bicho de 7 cabeças, obtemos 97% de precisão nos testes de previsão para as flores

### Para melhor visualização, código inteiro fica da seguinte maneira:

```python
#Bibliotecas usadas:
from sklearn.datasets import load_iris 
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier

import numpy as np
import pandas as pd




X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0) #separando as variáveis entre treino/teste

knn = KNeighborsClassifier(n_neighbors=1) #criando o objeto knn no modelo KNN de Machine Learning
knn.fit(X_train, y_train) #treinando o modelo no objeto knn
    
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
